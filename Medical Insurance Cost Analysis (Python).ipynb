# Medical Insurance Cost Dataset Analysis 

Investigation of the Medical Insurance dataset from Kaggle using python functions and classes.

```python
import csv
```

create empty lists for different columns of the dataset

```python
age = []
sex= []
bmi=[]
children = []
smoker=[]
region =[]
charges= []
```

read csv file and convert to multiple lists

```python
with open('insurance.csv') as file:
    output = csv.DictReader(file)
    for items in output:
        age.append(items['age'])
        sex.append(items['sex'])
        bmi.append(items['bmi'])
        children.append(items['children'])
        smoker.append(items['smoker'])
        region.append(items['region'])
        charges.append(items['charges'])
```

check if the data has been saved to the individual lists

```python
smoker
print(len(smoker))
```

We have 1,338 rows/observations in the dataset.

calculate the average of certain parameters from the dataset

create a function to calculate averages

```python
def avg(lst):
    sum_items = 0
    for items in lst:
        sum_items += float(items)
    items_average = sum_items/len(lst)
    return items_average
```

### Age average

```python
print(avg(age))
```

Average age of our dataset is 39 years

### Charges Average

```python
print(avg(charges))
```

The average insurance charge paid is 13,270
